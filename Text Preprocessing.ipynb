# Example why lemmatization is better then stemmer


```python
from nltk.stem.wordnet import WordNetLemmatizer
lem = WordNetLemmatizer()

from nltk.stem.porter import PorterStemmer
stem = PorterStemmer()

word = "flying"
print("Lemmatized Word:",lem.lemmatize(word,"v"))
print("Stemmed Word:",stem.stem(word))
```

# Coreference resolution (anaphora resolution)

.concordance() is a special nltk function. So you can't just call it on any python object (like your list).

More specifically: .concordance() is a method in the Text class of nltk

Basically, if you want to use the .concordance(), you have to instantiate a Text object first, and then call it on that object.

# concordance() is a method of the Text class of NLTK. It finds words and displays a context window. Word matching is not case-sensitive.
# concordance() is defined as follows: concordance(self, word, width=79, lines=25). Note default values for optional params.

# def collocations(self, num=20, window_size=2). num is the max no. of collocations to print.

```python
#Collocation extraction
from nltk.text import Text
#whitespace_removal.collocations()
text=Text(tokens)
text.concordance('first')
text.collocations()
```

```python
#Relationship extraction
text.collocations()
text.similar('brother') # # similar(self, word, num=20). Distributional similarity: 
#find other words which appear in the same contexts as the specified word; list most similar words first. 
#if guy was there then it would have searched
text.dispersion_plot(['man', 'thing', 'quack'])  ## Reveals patterns in word positions. Each stripe represents an instance of a word, and each row represents the entire text.
text.plot(20)   # # plots 20 most common tokens
```
